```python
import autokeras
import pandas as pd
from sklearn.model_selection import train_test_split


from sklearn import metrics
```

```python
file_name = "/course/wine.csv"
wine = pd.read_csv(file_name)
wine.head()
```

```python
file_name = "/course/insurance.csv"
insurance = pd.read_csv(file_name)
insurance.head()
```

```python
#The main difference between Regression and Classification algorithms that Regression algorithms 
#are used to predict the continuous values such as price, salary, age, etc. and 
#Classification algorithms are used to predict/Classify the discrete values such as 
#Male or Female, True or False, Spam or Not Spam, etc.
```

```python
#recordar que las NNR pueden utilizar solo valores numericos
#a autokeras no le importa, el hace todo por nosotors


x_train,x_test,y_train,y_test = train_test_split(insurance.iloc[:,:-1],insurance.iloc[:,-1])
```

```python
#tenemos un problema de regression ...lineal
keras = autokeras.StructuredDataRegressor(max_trials=8)
#van por el training data epochs veces
keras.fit(x_train, y_train, epochs=10)

predictions = keras.predict(x_test)

#en keras se almacenan todos los hyperparametos mejorados
```

```python

metrics.mean_squared_error(y_test,predictions)
```

```python
predictions = keras.predict(x_test)
```

```python
#un problema de clasificaciones
x_train,x_test,y_train,y_test = train_test_split(wine.iloc[:,:-1],wine.iloc[:,-1])
```

```python

keras = autokeras.StructuredDataClassifier(max_trials=8)
# Fit the training dataset
keras.fit(x_train, y_train, epochs=100)
```

```python
# Predict the validation data
pred_keras = keras.predict(x_test)
# Compute the accuracy
print('Accuracy: ' + str(metrics.accuracy_score(y_test, pred_keras)))
```
