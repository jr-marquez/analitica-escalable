# Utilizar Dask para Clasificación

```python
from dask_ml.cluster import KMeans
import dask_ml.datasets
```

## Puedes crear datasets más grandes que la memoria del equipo

```python
X, y = dask_ml.datasets.make_blobs(n_samples=500000,
                                   chunks=50000,
                                   random_state=0,
                                   centers=5)
```

```python
clf = KMeans(n_clusters=5,
            init_max_iter=20)
clf.fit(X)
```

```python
my_pred = clf.predict(X[:10])
```

```python
my_pred.compute()
```
